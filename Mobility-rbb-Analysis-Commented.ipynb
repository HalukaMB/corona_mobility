```python
#here we define three crucial functions

def avg_extactor(dfname):
    #to get a list of the columnnames with the average in the name for the given dataframe
    avg_cols = [col for col in dfname.columns if 'Avg' in col]
    #we add the locationId
    avg_cols.extend(['locationId'])
    #and then we create this average dataframe by filtering
    avg_df=dfname[avg_cols]
    #and return the df but also the name of the columns
    return avg_df, avg_cols
```

```python
def current_extractor(dfname):
    #this is the exact opposite, we filte for all columnames without "Avg"
    current_cols = [col for col in dfname.columns if 'Avg' not in col]
    #and create a dataframe of the current time from this
    current_df=dfname[current_cols]
    #and return the df but also the name of the columns

    return current_df,current_cols
```

```python
def day_by_day(dfname,since_when,df_column):
    #we take in a dataframe and ask for the name of the column
    #to create a dataframe that only contains this name at the start
    time_df=pd.DataFrame({"name":[df_column]})
    
    #then we get from this dataframe only the current data without the past averages and the column names
    current_df_specs=current_extractor(dfname)
    current_cols=current_df_specs[1]
    current_df=current_df_specs[0]
    
    #and we get the averages without the current data
    avg_df_specs=avg_extactor(dfname)
    avg_cols=avg_df_specs[1]
    avg_df=avg_df_specs[0]

    #now we iterate through the list of columnnames from the current df
    #these are different days for which we know the number of movements
    for day in current_cols[since_when:]:
        #we retrieve the date
        date=pd.to_datetime(day,format="%Y-%m-%d")
        print(date)
        #and get the name of the weekday
        weekday=date.day_name()
        print(weekday)
        #now we find in our dataframe of past averages the corresponding column name with this weekday
        avg_col_def = [col for col in avg_df.columns if weekday  in col]
        
        #so that we have the sum of values for that day in the average df
        print(avg_df[avg_col_def[0]].sum())
        
        #and that we have the sum of values for that day in the current df
        print(current_df[day].sum())
        
        #we calculate the ratio and save it in the dataframe under the columnname of that day
        time_df[day]=(current_df[day].sum()/avg_df[avg_col_def[0]].sum()*100)-100
    
    #we transpose
    time_df=time_df.transpose()
    
    #take the name from the first row as the columnames
    time_df.columns = time_df.iloc[0]
    #then drop this row, reset the indext and return the df
    time_df=time_df.drop(time_df.index[0])
    time_df=time_df.reset_index()
    return time_df
    
```

```python
germandf=day_by_day(total_germany_data,10,"Deutschland")
```

```python
#and then we do the same thing but for sums of regions
def weekchart(dfname,region):
    #we take in the dataframe for the region and the name of the region

    #we get for the region all the current values 
    current_df_specs=current_extractor(dfname)
    current_cols=current_df_specs[1]
    current_df=current_df_specs[0]
    
    #and we get for the region all past avg values
    avg_df_specs=avg_extactor(dfname)
    avg_cols=avg_df_specs[1]
    avg_df=avg_df_specs[0]
    
    #then we save the name of the region
    weekdict={"name":[region]}
    
    #and we iterate through the last weeks
    for weeknumber in range(0,numberofweeks,1):
        # for each week we create a dataframe
        lastweeks=pd.DataFrame()
        lastweeks["name"]=[region]
        lastweeks
        print(weeknumber)
        #get the first day of the week
        firstday=timespan_cols[weeknumber*7]
        
        #and then we iterate day by day
        for day in timespan_cols[weeknumber*7:(weeknumber+1)*7]:
            
            #if the day is in the holidays-list, we skip
            if (day) in (holidays):
                pass
            else:
                #get the name of the weekday
                print(day)
                date=pd.to_datetime(day,format="%Y-%m-%d")
                weekday=date.day_name()
                print(weekday)

                #and get from the 2019 avg dataframe the columnname for the corresponding day
                avg_col_def = [col for col in avg_df.columns if weekday in col]
                
                print(current_df[day])
                print(avg_df[avg_col_def[0]])

                
                #then we get the ratio of movement on that day in reference to the 2019 avg
                #by summing all movements of that current day over the dataframe
                #and summing all movement of that weekday over the avg df 
                #and then calculating the ratio
                lastweeks[weekday]=(current_df[day].sum()/avg_df[avg_col_def[0]].sum()*100)-100
        
        #and end up with this dataframe of which again we calculate the mean
        print(lastweeks)
        listofcolumns=lastweeks.columns.tolist()
        mean= lastweeks[listofcolumns[2:]].mean(axis=1)
        print(mean)
        
        #and save this mean to the dictionary
        weekdict[firstday]=mean.tolist()
    
    #and we built our dictionary from there
    regionmeandf=pd.DataFrame(weekdict)
    return(regionmeandf)
```

```python
berlin_per_week=weekchart(berlin_data,"Berlin")
brandenburg_per_week=weekchart(brandenburg_data,"Brandenburg")
germany_per_week=weekchart(total_germany_data,"Deutschland")
germany_per_week
```
